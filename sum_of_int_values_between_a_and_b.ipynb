```python
#Վարժություն 1.6
""" Ստեղծել ֆունկցիա, որը ստանում է երկու թվային արգումենտ` a և b, և
վերադարձնում է a-ից մինչև b ընկած ամբողջ թվերի գումարը։ Կարող եք
ենթադրել, որ առաջին արգումենտը միշտ փոքր է երկրորդից։"""

def sum_of_int_values_between_a_and_b(a,b):
    result=0
    while a<b:
        result +=a
        a +=1
    return result
sum_of_int_values_between_a_and_b(7,10)
```
